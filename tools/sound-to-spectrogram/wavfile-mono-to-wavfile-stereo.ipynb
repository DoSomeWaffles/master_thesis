```python
# Import libraries
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import os
import numpy as np
# for visualizing the data
import matplotlib.pyplot as plt
import zipfile
```

```python
def dual_mono_to_single_stereo(file):
    try:
        print(file)
        if file.endswith('1.wav'):
            
            Fs1, aud1 = wavfile.read(file)
            Fs2, aud2 = wavfile.read(file[:-5]+'2.wav')

            aud1_cropped = crop_audio(Fs1,aud1)
            aud2_cropped = crop_audio(Fs2,aud2)

            stereo_audio = np.array([aud1_cropped, aud2_cropped]).T
            print(file)
            wavfile.write(file[:-5]+'stereo.wav', Fs1, stereo_audio)
    except:
        print(file)

def crop_audio(Fs,aud):
    noise_value = aud.max()
    noise_index = np.where(aud == noise_value)
    after_noise_crop = 0.20
    clip_length = 2
    aud_cropped = aud[noise_index[0][0]+int(after_noise_crop*Fs):noise_index[0][0]+clip_length*Fs]
    return aud_cropped
```

```python


# get the zip path
zip_path = '../../sound-propagation-simulation/dataset/2023-04-03 10-46-21.zip'

# unzip the files
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('../../sound-propagation-simulation/dataset/')

# Get the folder path
folder = zip_path[:-4]
# List the files in the folder and every subfolder
file_list =[]
for path, subdirs, files in os.walk(folder):
    for name in files:
        if name.endswith('.wav'):
            file_list.append(os.path.join(path, name))
print(file_list)
```

```python
# run wav_to_img function for each file in parallel
from tqdm import tqdm
from joblib import Parallel, delayed
for file in file_list:
    dual_mono_to_single_stereo(file)
    print(file)
#Parallel(n_jobs=4,verbose=10)(delayed(dual_mono_to_single_stereo)(files[i]) for i in range(len(files)))
```

```python
# prepave validation dataset
# for each file in the validation folder

validation_folder = '../../sound-propagation-simulation/dataset/validation/'
print(validation_folder)
# print current folder
print(os.getcwd())
# List the files in the folder and every subfolder
file_list =[]
for path, subdirs, files in os.walk(validation_folder):
    for name in files:
        if name.endswith('.wav'):
            file_list.append(os.path.join(path, name))
print(file_list)
# run crop_audio function for each file 
for file in file_list:
    # load the audio
    Fs1, audio = wavfile.read(file)
    aud_cropped = crop_audio(Fs1, audio)
    print(file)
    # save the cropped audio
    wavfile.write(file[:-4]+'_cropped.wav', Fs1, aud_cropped)
    


```
