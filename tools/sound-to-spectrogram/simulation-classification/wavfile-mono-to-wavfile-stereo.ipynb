```python
# Import libraries
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import os
import numpy as np
# for visualizing the data
import matplotlib.pyplot as plt
import zipfile
from tqdm import tqdm
from joblib import Parallel, delayed
```

```python
def dual_mono_to_single_stereo(file):
    try:
        if file.endswith('1.wav'):
            
            Fs1, aud1 = wavfile.read(file)
            Fs2, aud2 = wavfile.read(file[:-5]+'2.wav')

            aud1_cropped = crop_audio(Fs1,aud1)
            aud2_cropped = crop_audio(Fs2,aud2)

            stereo_audio = np.array([aud1_cropped, aud2_cropped]).T
            wavfile.write(file[:-5]+'stereo.wav', Fs1, stereo_audio)
    except Exception as e:
        print(e)
        pass
    
def crop_audio(Fs,aud):
    noise_value = aud.max()
    noise_index = np.where(aud == noise_value)
    after_noise_crop = 0.20
    clip_length = 2
    aud_cropped = aud[noise_index[0][0]+int(after_noise_crop*Fs):noise_index[0][0]+clip_length*Fs]
    return aud_cropped
```

```python
def unzip(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(zip_path[:zip_path.rfind('/')+1])
        file_list =[]
        folder = zip_path[:-4]
        for path, subdirs, files in os.walk(folder):
            for name in files:
                if name.endswith('.wav'):
                    file_list.append(os.path.join(path, name))
        print(file_list)
        for file in tqdm(range(len(file_list))):
            dual_mono_to_single_stereo(file_list[file])
    print(zip_path,'Done')
```
