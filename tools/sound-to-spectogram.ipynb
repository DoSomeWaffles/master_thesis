```python
# for data transformation
import numpy as np
# for visualizing the data
import matplotlib.pyplot as plt
# for opening the media file
import scipy.io.wavfile as wavfile
```

```python
Fs1, aud1 = wavfile.read('ambulance.wav')
Fs2, aud2 = wavfile.read('firetruck1.wav')
Fs3, aud3 = wavfile.read('firetruck2.wav')
Fs4, aud4 = wavfile.read('riot.wav')
# select left channel only
aud1 = aud1[:,0]
aud2 = aud2[:,0]
aud3 = aud3[:,0]
aud4 = aud4[:,0]
# trim the first 125 seconds
first = aud1[:int(Fs1*100)]
second = aud2[:int(Fs2*100)]
third = aud3[:int(Fs3*100)]
fourth = aud4[:int(Fs4*300)]
```

```python
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(first, Fs=Fs1)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()

powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(second, Fs=Fs2)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()

powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(third, Fs=Fs3)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()

powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(fourth, Fs=Fs4)
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.show()
```
